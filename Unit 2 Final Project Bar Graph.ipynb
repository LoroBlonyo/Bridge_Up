```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salinity_info_38N_68W.csv")
```

```python
print(df.head())
```

```python
depth = df["Depth "]
temp = df["Temperature (Celsius)"]
sal = df["Salinity (PSU)"]
den = df["Density (kg/m3)"]
```

```python
# Hint: You need to specify the x data and the y data for the bar graph
plt.plot(depth, temp)
plt.plot(depth, sal)

plt.title("Temperature and Salinity")
plt.xlabel("Depth")

plt.legend()

plt.show()
```

```python
# Hint: You need to specify the x data and the y data for the bar graph
plt.plot(depth, temp)
plt.legend(loc=2)

plt.twinx()

plt.plot(depth, den)
plt.title("Temperature and Density")
plt.xlabel("depth")

plt.legend(loc = "center right")

plt.show()
```

```python
# This funcion clears the figure of any lines that were previously there. 
plt.clf()

# We specified the X and Y labels, and set the location of the legend
plt.plot(depth, temp, color="green")
plt.legend(loc=2)

# We named the X and the Y axes. 
plt.ylabel("Temperature(In Celsius)")
plt.xlabel("Depth")

# This function allows us to combine 2 different graphs that share the same X axis, but different Y axis.
plt.twinx()

# We again specified the X and Y labels, and set the location of the legend.
plt.plot(depth, sal, color="blue")
plt.legend()
# We set the title of the bar graph.
plt.title("Salinity and Temperature")

# We set the range of the graph
plt.xlim(0, 4000)
plt.ylim(0,40)

# We labeled the other Y ases. 
plt.ylabel("Salinity(psu)")
# The specifies where the tick marks will be. 
plt.xticks(range(0, 4500,500))
# This function is like the print function for graphs 
plt.show()
```
